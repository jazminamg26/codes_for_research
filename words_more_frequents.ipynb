```python
import re
from collections import Counter
import pandas as pd

def get_ngram_frequencies(sentences, n=1):
    text = " ".join(sentences)
    text = text.upper()
    text = re.sub(r'[^A-Z\s]', '', text)
    
    words = text.split()
    
    if n > 1:
        ngrams = zip(*[words[i:] for i in range(n)])
        tokens = [" ".join(ngram) for ngram in ngrams]
    else:
        tokens = words
        
    frequency_counts = Counter(tokens)
    
    df = pd.DataFrame(frequency_counts.items(), columns=['N-gram', 'Frequency'])
    df = df.sort_values(by='Frequency', ascending=False).reset_index(drop=True)
    
    return df
```
